#####_ord()_ es una función incorporada en Python 3, para convertir la cadena que representa un carácter Unicode en un entero que representa el código Unicode del carácter.
######Este diccionario se crea para tomar la ',' y eliminarla del dataset para poder convertir a datos númericos

```python
mi_diccionario = {ord(','): None }
```

#####El método _**translate()**_ devuelve una cadena en la que algunos caracteres especificados se sustituyen por el carácter descrito en un diccionario, o en una tabla de asignación.
#####El método _**astype()**_ de Python nos permite establecer o convertir el tipo de datos de una columna de datos existente en un conjunto de datos o un marco de datos. De este modo, podemos cambiar o transformar el tipo de los valores de los datos o de las columnas individuales o múltiples a otra forma utilizando la función _**astype()**_.

######Se indica que en las columnas del dataset se realicen modificaciones en estas, puesto que debemos cambiarlas de tipo object a tipo int.

```python
df['enero'] = df['enero'].str.translate(mi_diccionario).astype(int)
df['febrero'] = df['febrero'].str.translate(mi_diccionario).astype(int)
df['marzo'] = df['marzo'].str.translate(mi_diccionario).astype(int)
df['abril'] = df['abril'].str.translate(mi_diccionario).astype(int)
df['mayo'] = df['mayo'].str.translate(mi_diccionario).astype(int)
df['junio'] = df['junio'].str.translate(mi_diccionario).astype(int)
df['julio'] = df['julio'].str.translate(mi_diccionario).astype(int)
df['agosto'] = df['agosto'].str.translate(mi_diccionario).astype(int)
df['septiembre'] = df['septiembre'].str.translate(mi_diccionario).astype(int)
df['octubre'] = df['octubre'].str.translate(mi_diccionario).astype(int)
df['noviembre'] = df['noviembre'].str.translate(mi_diccionario).astype(int)
df['diciembre'] = df['diciembre'].str.translate(mi_diccionario).astype(int)
```

######Se indica que en las columnas del dataset se realicen modificaciones en estas, puesto que debemos cambiarlas de tipo object a tipo int y eliminar la ',' de estas.

```python
df['ExtranjeroEnero'] = df['ExtranjeroEnero'].astype(str).str.translate(mi_diccionario).astype(int)
df['ExtranjeroFebrero'] = df['ExtranjeroFebrero'].astype(str).str.translate(mi_diccionario).astype(int)
df['ExtranjeroMarzo'] = df['ExtranjeroMarzo'].astype(str).str.translate(mi_diccionario).astype(int)
df['ExtranjeroAbril'] = df['ExtranjeroAbril'].astype(str).str.translate(mi_diccionario).astype(int)
df['ExtranjeroMayo'] = df['ExtranjeroMayo'].astype(str).str.translate(mi_diccionario).astype(int)
df['ExtranjeroJunio'] = df['ExtranjeroJunio'].astype(str).str.translate(mi_diccionario).astype(int)
df['ExtranjeroJulio'] = df['ExtranjeroJulio'].astype(str).str.translate(mi_diccionario).astype(int)
df['ExtranjeroAgosto'] = df['ExtranjeroAgosto'].astype(str).str.translate(mi_diccionario).astype(int)
df['ExtranjeroSeptiembre'] = df['ExtranjeroSeptiembre'].astype(str).str.translate(mi_diccionario).astype(int)
df['ExtranjeroOctubre'] = df['ExtranjeroOctubre'].astype(str).str.translate(mi_diccionario).astype(int)
df['ExtranjeroNoviembre'] = df['ExtranjeroNoviembre'].astype(str).str.translate(mi_diccionario).astype(int)
df['ExtranjeroDiciembre'] = df['ExtranjeroDiciembre'].astype(str).str.translate(mi_diccionario).astype(int)
```

#####Se realiza la suma de los datos de visitantes extranjeros y nacionales de cada mes en una columna nueva con base al mes sumado.

####La biblioteca de gráficos de Python de Plotly hace gráficos interactivos de calidad de publicación. Ejemplos de cómo hacer gráficos de líneas, gráficos de dispersión, gráficos de área, gráficos de barras, barras de error, gráficos de caja, histogramas, mapas de calor, subtramas, ejes múltiples, gráficos polares y gráficos de burbujas.

```python
fig = px.histogram(df3, x="Centro INAH", y="Enero", color="Estado", marginal="rug",
                   hover_data=df3.columns, width=1000, height=1300,
                   labels={'Enero': 'Total'})
fig.update_layout(xaxis_title = "Recintos Culturales",
                  yaxis_title = "Total de visitantes",
                  title={
                    'text': "Visitantes nacionales y extranjeros, a los Recintos Culturales en México durante 2022",
                    'y':0.97,
                    'x':0.5,
                    'xanchor':'center',
                    'yanchor':'top'
                  })
fig.show()
```
